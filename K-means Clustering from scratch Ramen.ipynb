```python
idx = np.random.randint(m, size=3) # it generates random 3 number from integer > 1 to m 
C=d1[idx]# we select those three positional data points as our initial centroids
```

```python
 # Function for distance measurement of each point from each centroids
def dist_cal(C,d1): # Note centroid matrix C and data d1 as input argument
    c1=C[0,:]; c2=C[1,:] ;c3=C[2,:] ##Seperating individual three centroids
    c1d=np.sqrt(np.sum((d1-c1)**2,axis=1))
    c1d=c1d[:,np.newaxis]
    c2d=np.sqrt(np.sum((d1-c2)**2,axis=1))
    c2d=c2d[:,np.newaxis]
    c3d=np.sqrt(np.sum((d1-c3)**2,axis=1))
    c3d=c3d[:,np.newaxis]
    D=np.hstack((c1d,c2d,c3d))
    return D
```

```python
## Function for group assignment 
def group_split(m,d1,D):
    group=np.zeros((m,1))
    for i in range(m):
        a=D[i,0]
        b=D[i,1]
        c=D[i,2]
        if ( a < (b and c) ):
            group[i,0]=1
        elif ( b < (a and c) ):
            group[i,0]=2
        else:
            group[i,0]=3  
    d_group=np.hstack((d1,group)) # adding group label as a column 
    return d_group
```

```python
for i in range(100):
    D=dist_cal(C,d1) ## Distance function call
    d_with_group=group_split(m,d1,D) ## group function call
    
    ## Splitting the three different group
    dg1=d_with_group[d_with_group[:,2]==1]
    dg2=d_with_group[d_with_group[:,2]==2]
    dg3=d_with_group[d_with_group[:,2]==3]
    ## Mean calculation and assignmet of new centroids
    C= np.array([np.mean(dg1,axis=0),np.mean(dg2,axis=0),np.mean(dg3,axis=0)])
    C= C[:,:2]
C   
```
