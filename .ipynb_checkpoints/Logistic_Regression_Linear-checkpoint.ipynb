```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

### Data Loading
* For sake of simplicity,sklearn is used below to make linearly seperable data set. You can use any linearly seperable dataset.

```python
from sklearn.datasets import make_blobs

(X,y) =  make_blobs(n_samples=100,n_features=2,centers=2,cluster_std=1.05,random_state=20)
# we need to add 1 to X values (Vectorization)
X1 = np.c_[np.ones((X.shape[0])),X]

plt.scatter(X1[:,1],X1[:,2],marker='o',c=y)
plt.show()
```

```python
print(X1[:5])    # Displaying first five rows of X1
print(X1.shape)  # Displaying shape of X1
print(y[:5])     # Displaying first five member  of y, note is one dimensional array
```

```python
X=X1.T #Transposing X1 and saving into X again 
print(X.shape)
X.shape[0]
## Note: Here, X has three features and total 100 set of data
```

```python
# Initializing parameter or weights vector theta
np.random.seed(0)
theta=np.random.randn(X.shape[0],1) # (3,1) # After transposing--> (1,3)
theta
```

#### Making sigmoid function and plotting it

```python
def sigmoid(z):
    return (1 / (1 + np.exp(-z)))
```

```python
x1=np.linspace(-10,10,200)
y1=sigmoid(x1)

plt.plot(x1,y1,"g--",linewidth=2,label='Sigmoid')
plt.grid(True)
plt.axvline(c='r')
plt.axhline(c='r')
plt.legend(fontsize=14)
plt.title('Sigmoid Function',fontsize=14)
plt.axis([-10,10,-0.5,1.2])
plt.show()
```

### Gradient calculation
* Understanding the implementation step-wise and then making function for calculating gradient.
* Equation below is Gradient Descent and Last term of RHS is the Gradient of cost function see derivation in the slide.


* calculating predicted y by passing,theta.T.X (theta transpose x) into sigmoid

```python
yp=sigmoid((theta.T).dot(X))  
print(yp.shape)
print(yp)
```

```python
y[np.newaxis,:] # conveting given y into 2D
```

```python
## Calculating --> yp-y
(yp-y[np.newaxis,:])
```

```python
# Finally gradient calculation by Matrix product of X and (yp-y)
(X).dot((yp-y[np.newaxis,:]).T)
```

```python
# This will also work
(X).dot((yp-y).T)
# Note transpose operation is often done to match the order of two matrix for matrix multiplication
```

#### Finally making function for gradient calculation:

```python
def gradient_cal(theta,X,y): ## theta --> (3,1),# X-->(3,100), #y-->(100,)
    yp=sigmoid((theta.T).dot(X))
    grad=X.dot((yp-y).T)
    return grad
```

### Gradient descent Application

```python
np.random.seed(0) # For fixing the initial randomized theta
theta=np.random.randn(X.shape[0],1)
print(theta) ## Initial theta
alpha=0.001
for i in range(100000):
    grad=gradient_cal(theta,X,y)
    theta=theta-alpha*grad
theta  ## Final updated theta after 100000 iteration
```

### Including cost function


```python
##  Implementing above eqn into syntax
m=len(y)
cost=(-1/m)*np.sum(y*np.log(yp)+(1-y)*np.log(1-yp))
cost
```

```python
theta=np.random.randn(X.shape[0],1)
alpha=0.001
iteration=100000
cost=np.zeros(iteration)  ## Additional Part
m=len(y)
for i in range(iteration):
    grad=gradient_cal(theta,X,y)
    theta=theta-alpha*grad
    yp=sigmoid((theta.T).dot(X))  ## Additional Part
    cost[i]=(-1/m)*np.sum(y*np.log(yp)+(1-y)*np.log(1-yp))  
    #print(cost[i])
theta
```

```python
plt.plot(cost)
```

#### PLOTTING Decision Boundary

```python
c=theta[0,0]
a=theta[1,0]
b=theta[2,0]
## Convertion ax+by+c=0 into y=mx+c
Y2=(-c-a*X[1,:])/b
plt.plot(X[1,:],Y2)
######## Plotting given data
plt.scatter(X1[:,1],X1[:,2],c=y)
plt.title('Visualization of Data',fontsize=12,style='italic',fontweight='bold')
plt.xlabel('X1',fontsize=12)
plt.ylabel('X2',fontsize=12)
plt.show()
```

#### Making Prediction Function
* Now we can predict the output for each given X using updated theta.

```python
 X
```

```python
yp=sigmoid((theta.T).dot(X))
yp
```

```python
def predict(X,theta):       ## X-->(3,100)  ,theta-->(3,1)
    yp=sigmoid((theta.T).dot(X))
    yp=np.where(yp>0.5,1,0)    
    return yp
```

```python
label_pred=predict(X,theta) # This will give us the predicted labels for each X.
label_pred # Resul is 2D array with one row and multiple columnn
```

```python
y_pred=label_pred[0] # Selecting row to convering into 1D
y_pred
```

```python
# actual y
y  
```

### Calculating Accuracy
* We can calculate accuracy by comparing actual y(Initial given in table for each X) and predicted y  from our model
* Implementing step-wise.

```python
y_pred!=y # error 
```

```python
#error and acuracy 
error=np.mean(y_pred!=y)
print(error)
accuracy=(1-error)
accuracy
```
